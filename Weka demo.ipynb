```python
f = open("foods.txt","r")

L = []
oneLine = []
allWords = []
W = []
listOfReviews = []

for i in range(10000):
    x = f.readline()
    if ("review/text:" in x):
        x = x[13:]
        x = x.lower()
        x = x.replace(".", "")
        x = x.replace("\n", "")
        x = x.replace(",", "")
        x = x.replace("!", "")
        x = x.replace("<br>", "")
        x = x.replace("<br/>", "")
        x = x.replace("<br />", "")
        x = x.replace("(", "")
        x = x.replace(")", "")
        oneLine = x.split(" ")
        listOfReviews.append(x)
        
        for i in oneLine:
            allWords.append(i)
            if i not in L:
                L.append(i)
                
f.close()

print(listOfReviews)
```

```python
f1 = open("stopwords.txt", "r")

stopwords = []

for i in range(665):
    x = f1.readline()
    x = x.replace("\n", "")
    stopwords.append(x)
```

```python
W = L
for i in stopwords:
    if i in W:
        W.remove(i)

print(W)
```

```python
vector = {}

for i in W:
    count = allWords.count(i)
    vector[i] = count
    
print(vector)
```
